```python
# import 

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Diff between CDF and PDF

# CDF: Cumulative Distribution Function

# PDF: Probability Density Function

# CDF: The probability that the variable takes a value less than or equal to x

# PDF: The probability that the variable takes a value exactly equal to x

```

```python
# CDF

x = np.random.normal(0, 1, 1000)
pdf = stats.norm.pdf(x)
cdf = stats.norm.cdf(x)


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title('PDF')
plt.subplot(1, 2, 2)
plt.plot(x, cdf)
plt.title('CDF')
plt.show()
```
