```python
#libs
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Normal Distribution
data = np.random.normal(0, 1, 1000)

# Plot
plt.figure(figsize=(10, 5))
sns.histplot(data, kde=True)
plt.title('Normal Distribution')
plt.show()
```

```python
# example
# 68% of the data is within 1 standard deviation of the mean

# 1 standard deviation
std = 1
data = np.random.normal(0, std, 1000)

# Plot
plt.figure(figsize=(10, 5))
sns.histplot(data, kde=True)
plt.title('Normal Distribution')
plt.show()
```

```python
# example 2 One factory product 1000 items per day with a mean of 100 and a standard deviation of 15 items per day. if the factory produces 1000 items per day, how many items are expected to be between 85 and 115 items per day?

mean = 100
std = 15

# 85 items

z1 = (85 - mean) / std

# 115 items

z2 = (115 - mean) / std

# P(85 < x < 115)

p = stats.norm.cdf(z2) - stats.norm.cdf(z1)

# Expected value

p * 1000


```

```python
# example 3 a average of 4300 terrains are destroyed by a fire per year. std is 562500 terrains. what is the probability that the number of terrains destroyed by fire in any year is between 2500 and 4200?

mean = 4300
var = 562500

distribution = stats.norm(mean, np.sqrt(var))

# 2500 terrains

prob1 = distribution.cdf(2500)

# 4200 terrains

prob2 = distribution.cdf(4200)

# P(2500 < x < 4200)

# plot distribution
x = np.linspace(0, 10000, 1000)
y = distribution.pdf(x)

plt.figure(figsize=(10, 5))

plt.plot(x, y)
plt.fill_between(x, y, where=(x > 2500) & (x < 4200), color='red', alpha=0.5)
plt.title('Normal Distribution')
plt.show()





```

```python
# Example 5 : A packed-service restaurant has a mean of 30 min and variance is 25 min. What is the prob that a costumer will wait between 20 and 40 min?

mean = 30
var = 25

distribution = stats.norm(mean, np.sqrt(var))

# 20 min

prob_1 = distribution.cdf(20)

# 40 min

prob_2 = distribution.cdf(40)

# P(20 < x < 40)

prob = prob_2 - prob_1

prob

```
