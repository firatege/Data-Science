```python
# import

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# PMF : Probability Mass Function (확률질량함수)

# Summery of PMF : https://en.wikipedia.org/wiki/Probability_mass_function

x = np.array([1,2,3,4])
y = np.array([1/8,3/8,3/8,1/8])

plt.bar(x, y)
plt.show()


```

```python
# PDF : Probability Density Function (확률밀도함수)

# Summery of PDF : https://en.wikipedia.org/wiki/Probability_density_function

x = np.linspace(-5, 5, 1000)
pdf = stats.norm.pdf(x, loc=0, scale=1)

plt.plot(x, pdf)
plt.show()
```
