```python
#libs
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Discrete Uniform Distribution

# A discrete uniform distribution is a symmetric probability distribution whereby a finite number of values are equally likely to be observed; every one of n values has equal probability 1/n.

# The probability mass function of the discrete uniform distribution is:

# f(x) = 1/n for all x in {1, 2, 3, ..., n}

# where n is the number of values in the support.

# The mean of the discrete uniform distribution is:

# E[X] = (a + b) / 2

# where a is the smallest value in the support and b is the largest value in the support.

# The variance of the discrete uniform distribution is:

# Var(X) = (b - a + 1)^2 / 12

```

```python
# example

# Let's say we have a fair 6-sided die. The probability of each face is 1/6. This is an example of a discrete uniform distribution because each face has an equal probability of 1/6.

x = np.arange(1, 7)
y = np.repeat(1/6, 6)

plt.bar(x, y)

plt.xlabel('Number on Die')
plt.ylabel('Probability')
plt.title('Discrete Uniform Distribution')
plt.show()
```

```python
# Continuous Uniform Distribution

# A continuous uniform distribution is a symmetric probability distribution whereby a random variable has an equal probability of falling anywhere between two specified values.

# The probability density function of the continuous uniform distribution is:

# f(x) = 1 / (b - a) for a <= x <= b

# where a is the lower bound and b is the upper bound.

# The mean of the continuous uniform distribution is:

# E[X] = (a + b) / 2

# The variance of the continuous uniform distribution is:

# Var(X) = (b - a)^2 / 12
```

```python
# example 

# Let's say anybody can arrive at a bus stop between 7:00 and 7:30. The probability of arriving at any time between 7:00 and 7:30 is equal. 

x = np.linspace(7, 7.5, 100)

y = np.repeat(1/0.5, 100)

plt.plot(x, y)

plt.xlabel('Time')

plt.ylabel('Probability')

plt.title('Continuous Uniform Distribution')

plt.show()

```
