```python
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
import numpy as np

# Zar atışı sonuçlarını ortalama olarak kaydetmek için bir liste
mean_list = []

# x kadar deneme yap
x = 10000# İstediğiniz sayıda deneme yapabilirsiniz, burada örneğin 10 seçildi

for i in range(1, x + 1):
    # Her deneme için farklı rastgele zar atışları üret
    probability = np.random.randint(1, 7, size=i)  # Zar 1-6 arasında değerler alır
    mean_list.append(probability.mean())
    print(f'Number of attempts: {i} \t Sample mean: {np.mean(probability):.2f}')

# Tüm denemelerin ortalamasına bakalım
overall_mean = np.mean(mean_list)
print(f'Overall mean after {x} attempts: {overall_mean:.2f}')
```

```python
plt.figure(figsize=(10, 6))


plt.plot(np.arange(1, x+1), mean_list, label='Sample Mean')


plt.axhline(y=3.5, color='r', linestyle='--', label='y=3.5')

plt.xlabel('Number of Attempts')
plt.ylabel('Sample Mean')
plt.title('Sample Mean of Dice Rolls')
plt.legend()
plt.show()
```
