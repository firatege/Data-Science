```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# t-distiribution
np.random.seed(42)
data1 = np.random.standard_t(df=10, size=15)
data2 = np.random.standard_t(df=10, size=15)
data5 = np.random.standard_t(df=5, size=15)
data10 = np.random.standard_t(df=10, size=15)

```

```python
# histogram
plt.figure(figsize=(10, 6))
sns.kdeplot(data1, color='r', label='t-dist (df=10)')
sns.kdeplot(data2, color='b', label='t-dist (df=10)')
sns.kdeplot(data5, color='g', label='t-dist (df=5)')
sns.kdeplot(data10, color='y', label='t-dist (df=10)')
plt.legend()



```
