```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

```python
# trend types

# 1. linear trend

# 2. exponential trend

# 3. logarithmic trend

# 4. polynomial trend
```

```python
# 1. linear trend

x = np.arange(1, 100, 1)

y = 2*x + 1

plt.plot(x, y)
```

```python
# 2. exponential trend
x = np.arange(1, 100, 1)
def y(x):
    if x < 50:
        return 2**x
    else:
        return 2**x + 1000
    

y = [y(i) for i in x]

plt.plot(x, y)


```

```python
# 3. logarithmic trend

x = np.arange(1, 100, 1)

y = np.log(x)

plt.plot(x, y)
```

```python
# 4. polynomial trend

x = np.arange(1, 100, 1)

y = x**3

plt.plot(x, y)
```
